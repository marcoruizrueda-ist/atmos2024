**Step 1: Statistical Significance Tests**

We'll use the Pearson correlation test, which not only gives the correlation coefficient but also a p-value to determine statistical significance. A common threshold for significance is a p-value less than 0.05.

```python
# Statistical Significance Tests

import pandas as pd
import scipy.stats as stats
from scipy.stats import pearsonr

# Load the CSV file
file_path = 'methane_detection_papers.csv'
df_correlation = pd.read_csv(file_path)

# Rename columns to remove special characters
df_correlation.rename(columns={
    'Use ML?': 'Use_ML',
    'Is the dataset public?': 'Is_dataset_public',
    'Code available?': 'Code_available',
    'Reproducible?': 'Reproducible'
}, inplace=True)

# Convert categorical data to numerical form
df_correlation['Use_ML'] = df_correlation['Use_ML'].map({'Yes': 1, 'No': 0})
df_correlation['Is_dataset_public'] = df_correlation['Is_dataset_public'].map({'Yes': 1, 'No': 0, 'Yes with caveats': 0.5})
df_correlation['Code_available'] = df_correlation['Code_available'].map({'Yes': 1, 'No': 0})
df_correlation['Reproducible'] = df_correlation['Reproducible'].map({'Yes': 1, 'No': 0, 'Yes with caveats': 0.5})

# Extracting the 'Year' column which is numerical
numerical_cols = ['Use_ML', 'Is_dataset_public', 'Code_available', 'Reproducible', 'Year']

# Filter only numerical columns for correlation analysis
df_numerical = df_correlation[numerical_cols]

# Function to calculate Pearson correlation and p-value
def calculate_pearson_correlation(df):
    correlations = {}
    p_values = {}
    columns = df.columns
    for i in range(len(columns)):
        for j in range(i + 1, len(columns)):
            col1, col2 = columns[i], columns[j]
            corr, p_value = pearsonr(df[col1].dropna(), df[col2].dropna())
            correlations[(col1, col2)] = corr
            p_values[(col1, col2)] = p_value
    return correlations, p_values

# Calculate correlations and p-values
correlations, p_values = calculate_pearson_correlation(df_numerical)

# Display correlations and p-values
correlations, p_values
```

**Step 2: Regression Analysis**

We'll perform regression analysis to understand the relationships between the variables more deeply. This involves fitting linear regression models and examining their coefficients and significance.

```python
# Regression Analysis
import pandas as pd
import statsmodels.formula.api as smf

# Load the CSV file
file_path = 'methane_detection_papers.csv'
df_correlation = pd.read_csv(file_path)

# Rename columns to remove special characters
df_correlation.rename(columns={
    'Use ML?': 'Use_ML',
    'Is the dataset public?': 'Is_dataset_public',
    'Code available?': 'Code_available',
    'Reproducible?': 'Reproducible'
}, inplace=True)

# Convert categorical data to numerical form
df_correlation['Use_ML'] = df_correlation['Use_ML'].map({'Yes': 1, 'No': 0})
df_correlation['Is_dataset_public'] = df_correlation['Is_dataset_public'].map({'Yes': 1, 'No': 0, 'Yes with caveats': 0.5})
df_correlation['Code_available'] = df_correlation['Code_available'].map({'Yes': 1, 'No': 0})
df_correlation['Reproducible'] = df_correlation['Reproducible'].map({'Yes': 1, 'No': 0, 'Yes with caveats': 0.5})

# Fill missing values with the mean of the column
#df_correlation.fillna(df_correlation.mean(), inplace=True)

# Extracting the 'Year' column which is numerical
numerical_cols = ['Use_ML', 'Is_dataset_public', 'Code_available', 'Reproducible', 'Year']

# Filter only numerical columns for regression analysis
df_numerical_filled = df_correlation[numerical_cols]

# Perform regression analysis with 'Reproducible' as the dependent variable
model_reproducible = smf.ols(formula='Reproducible ~ Use_ML + Is_dataset_public + Code_available + Year', data=df_numerical_filled).fit()
summary_reproducible = model_reproducible.summary()

# Perform regression analysis with 'Is_dataset_public' as the dependent variable
model_public = smf.ols(formula='Is_dataset_public ~ Use_ML + Code_available + Reproducible + Year', data=df_numerical_filled).fit()
summary_public = model_public.summary()

# Display the regression results
print("=====================Regression with Reproducible as the Dependent Variable:=====================")
print(summary_reproducible)
print("\n=====================Regression with Is_dataset_public as the Dependent Variable:=====================")
print(summary_public)
```

#### **I know the number of papers and want to calculate the Power**

**Reverse Power Analysis: Calculating Power**

To calculate the statistical power given a pre-defined sample size (e.g., 35 papers), we need to set the effect size and alpha level. Let's assume we still expect a medium effect size (0.3) and use a standard alpha level of 0.05.

- **Effect Size:** A measure of the strength of the relationship. A medium effect size (0.3) is commonly used in social sciences.
- **Alpha (α) Level:** The probability of a Type I error, commonly set at 0.05.
- **Sample Size (nobs):** The number of observations (papers), which is 35 in this case.
- **Power:** The probability of correctly rejecting the null hypothesis (1 - Type II error).

Certainly! Here's a table summarizing the interpretation and analysis of using the specified parameters:

### Table: Interpretation and Analysis of Power Analysis Parameters

| **Parameter**   | **Value**            | **Interpretation**                                                                                                    |
|-----------------|----------------------|----------------------------------------------------------------------------------------------------------------------|
| **Effect Size** | 0.5                  | This indicates a large effect size, suggesting a strong relationship between the variables being studied.             |
| **Alpha (α)**   | 0.05                 | This is the standard significance level, indicating a 5% risk of a Type I error (rejecting a true null hypothesis).   |
| **Sample Size** | 35                   | The number of studies (papers) available for the analysis.                                                           |

### Table: Interpretation of Power Values

| **Estimated Power Range** | **Interpretation**                                                                                           |
|---------------------------|--------------------------------------------------------------------------------------------------------------|
| **0.90 and above**        | Excellent power: Very high probability of detecting a true effect. Low risk of Type II error.                |
| **0.80 - 0.89**           | Good power: High probability of detecting a true effect. Generally considered adequate in most studies.      |
| **0.70 - 0.79**           | Fair power: Moderate probability of detecting a true effect. Some risk of Type II error.                     |
| **0.60 - 0.69**           | Marginal power: Lower probability of detecting a true effect. Higher risk of Type II error.                  |
| **0.50 - 0.59**           | Low power: Low probability of detecting a true effect. Significant risk of Type II error.                    |
| **Below 0.50**            | Very low power: Very low probability of detecting a true effect. Very high risk of Type II error.             |

```python
from statsmodels.stats.power import NormalIndPower

# Define parameters
effect_size = 0.5  # Large effect size
alpha = 0.05  # Significance level
sample_size = 100  # Available sample size

# Perform power analysis to calculate power
power_analysis = NormalIndPower()

power = power_analysis.solve_power(effect_size=effect_size, alpha=alpha, nobs1=sample_size)
print(f'Estimated power for the study: {power:.2f}')
```
